```python
import pandas as pd

data= {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "June"],
    "Sales":[10000,12000,15000,13000,17000,16000],
    "Profit":[2000,3000,4000,2500,3500,3000]
}
df=pd.DataFrame(data)
print(df)
```

# 1.Line Plot of Monthy Sales


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,3))
plt.plot(df['Month'],df['Sales'],label='Sales',color='g',marker='o',linestyle='-')
plt.title('Sales Analysis over Months')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.legend()
plt.grid(True)
plt.show()
```

# 2.Bar Plot Months Vs Profit

```python
plt.figure(figsize=(6,3))
width=0.3
plt.bar(df['Month'],df['Sales'],label='Sales',color='b',width=width)
plt.bar(df['Month'],df['Profit'],label='Profit',color='g',width=width,bottom=df['Sales'])
plt.title('Sales and Profit Analysis over Months')
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 3.Pie Chart Profit VS Month


```python
from enum import auto
plt.figure(figsize=(8,5))
plt.pie(df['Profit'],labels=df['Month'],autopct='%1.1f%%',startangle=140,colors=plt.cm.Paired.colors)
plt.title('Profit Analysis by Month')
plt.show()
```

# 4.Scatter Plot

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Sales'],df['Profit'],label='Sales',color='b',s=100,edgecolors='black')
plt.title('Sales and Profit Analysis')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Histogram

```python
plt.figure(figsize=(8,4))
plt.hist(df['Sales'],bins=5,color='purple',edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

# 6.Box Plot

```python
plt.figure(figsize=(8,4))
plt.boxplot(df['Profit'],vert=False,patch_artist=True,boxprops=dict(facecolor='purple',color='black'))
plt.title('Profit Box Plot')
plt.xlabel('Profit')
plt.tight_layout()
plt.show()
```

```python
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt

data= {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "June"],
    "Sales":[10000,12000,15000,13000,17000,16000],
    "Profit":[2000,3000,4000,2500,3500,3000]
}
df=pd.DataFrame(data)

def generate_plot(plot_type):
  fig=plt.figure(figsize=(6,3))

  if plot_type=='Line Plot':
    plt.plot(df['Month'],df['Sales'],label='Sales',color='g',marker='o',linestyle='-')
    plt.title('Sales Analysis over Months')
    plt.xlabel('Month')
    plt.ylabel('Sales ($)')
    plt.legend()
    plt.grid(True)
  elif plot_type=='Bar Plot':
    fig.set_size_inches(10,6)
    width=0.3
    plt.bar(df['Month'],df['Sales'],label='Sales',color='b',width=width)
    plt.bar(df['Month'],df['Profit'],label='Profit',color='g',width=width,bottom=df['Sales'])
    plt.title('Sales and Profit Analysis over Months')
    plt.xlabel('Month')
    plt.ylabel('Amount ($)')
    plt.legend()
    plt.grid(True)
  elif plot_type=='Pie Chart':
    fig.set_size_inches(8,5)
    plt.pie(df['Profit'],labels=df['Month'],autopct='%1.1f%%',startangle=140,colors=plt.cm.Paired.colors)
    plt.title('Profit Analysis by Month')
  elif plot_type=='Scatter Plot':
    fig.set_size_inches(8,5)
    plt.scatter(df['Sales'],df['Profit'],label='Sales',color='b',s=100,edgecolors='black')
    plt.title('Sales and Profit Analysis')
    plt.xlabel('Sales')
    plt.ylabel('Profit')
    plt.legend()
    plt.grid(True)
  elif plot_type=='Histogram':
    fig.set_size_inches(8,4)
    plt.hist(df['Sales'],bins=5,color='purple',edgecolor='black')
    plt.title('Sales Distribution')
    plt.xlabel('Sales')
    plt.ylabel('Frequency')
  elif plot_type=='Box Plot':
    fig.set_size_inches(8,4)
    plt.boxplot(df['Profit'],vert=False,patch_artist=True,boxprops=dict(facecolor='purple',color='black'))
    plt.title('Profit Box Plot')
    plt.xlabel('Profit')

  plt.tight_layout()
  return fig
#gradio UI cod
demo=gr.Interface(
    fn=generate_plot,
    inputs=gr.Radio(choices=['Line Plot','Bar Plot','Pie Chart','Scatter Plot','Histogram','Box Plot']),
    outputs=gr.Plot(label="Sales Data Visualizations"),
    title='Sales & Proft Insights',
    description="Choose the type of graph for visualization"
)

demo.launch()


```
